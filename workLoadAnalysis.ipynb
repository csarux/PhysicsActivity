```python
import workLoadAnalysis as wl
import matplotlib.pyplot as plt
```

```python

# Filtro por patología. None: acumulada de todas las patologías, 'ORL', 'Pulmon', etc: solo esa patología
location=None
# Corrección por complejidad de la planificación. complexity=False para contabilizar el número de planificaciones únicamente.
complexity=True

acppadf = wl.create_physicist_activities(ppadf=wl.ppadf, physicists=wl.physicists, location=location, complexity=complexity)
fig, ax = plt.subplots(figsize=(12, 6))
if location:
    ax.set_title(f'Actividad acumulada normalizada por Físico para la localización {location}')
else:
    ax.set_title('Actividad acumulada normalizada por Físico para todas las localizaciones')
markers = ['o', 'D', 'v', '^', '<', '>', 'p', '*', 'X', 's']
for i, physicist in enumerate(wl.physicists):
    acppadf.plot(y=f'nac_{physicist}', ax=ax, marker=markers[i % len(markers)], markevery=5)
plt.show()
```
